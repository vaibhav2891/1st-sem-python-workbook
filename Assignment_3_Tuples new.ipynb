<h2>Assignment 3 : TUPLE

## Q1 Max and Min

```python
t = (3, 7, 1, 9, 4)
max_value = t[0]
min_value = t[0]
for i in t:
    if i > max_value:
        max_value = i
    if i < min_value:
        min_value = i
print("Max =", max_value)
print("Min =", min_value)
```

## Q2 Count occurrences

```python
t = (1, 2, 3, 2, 1, 3, 2)
freq = {}
for i in t:
    freq[i] = freq.get(i, 0) + 1
print(freq)
```

## Q3 Join tuple strings

```python
t=('Hello','World','Python')
result=''
for word in t:
    result+=word
print(result)
```

## Q4 Sum of even numbers

```python
t = (1, 2, 3, 4, 6)
sum_even = 0
for i in t:
    if i % 2 == 0:
        sum_even += i
print(sum_even)
```

<h2>2. Sort tuple of tuples by second element

```python
t = ((1, 5), (3, 2), (2, 8))
print(tuple(sorted(t, key=lambda x: x[1])))
```

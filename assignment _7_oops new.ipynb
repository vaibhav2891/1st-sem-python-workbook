```python
# Q.1) Create a class called Rectangle with attributes length and width. Add methods to calculate area and perimeter 

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


l = int(input("Enter length: "))
w = int(input("Enter width: "))
r = Rectangle(l, w)

print("Area of Rectangle:", r.area())
print("Perimeter of Rectangle:", r.perimeter())
```

```python
# Q.2) Create a class BankAccount with methods to deposit, withdraw, and display balance.

class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def show(self):
        print("Balance:", self.balance)

b = int(input("Enter balance: "))
a = BankAccount(b)

d = int(input("Enter deposit amount: "))
a.deposit(d)

w = int(input("Enter withdraw amount: "))
a.withdraw(w)

a.show()
```

```python
# Q.3) Create a parent class Vehicle and child classes Car and Bike that inherit from it. Override a method in child classes

class Vehicle:
    def move(self):
        print("Vehicle is moving")

class Car(Vehicle):
    def move(self):
        print("Car is moving on road")

class Bike(Vehicle):
    def move(self):
        print("Bike is moving on road")

v = Vehicle()
c = Car()
b = Bike()

v.move()
c.move()
b.move()
```

```python
# Q.4) Create a class Student with private attributes and public methods to access them.

class Student:
    def __init__(self, name, marks):
        self.__name = name
        self.__marks = marks

    def get_name(self):
        return self.__name

    def get_marks(self):
        return self.__marks

n = input("Enter student name: ")
m = int(input("Enter marks: "))
s = Student(n, m)

print("Name:", s.get_name())
print("Marks:", s.get_marks())
```

```python
# Q.5) Create a class Circle with method to calculate area. Overload the + operator to add two circles(add their radii)

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def __add__(self, other):
        return Circle(self.radius + other.radius)

r1 = int(input("Enter radius of first circle: "))
r2 = int(input("Enter radius of second circle: "))

c1 = Circle(r1)
c2 = Circle(r2)
c3 = c1 + c2

print("Area of first circle:", c1.area())
print("Area of second circle:", c2.area())
print("Radius after adding circles:", c3.radius)
print("Area of new circle:", c3.area())
```

```python
# Q.1) Create a class hierarchy: Animal -> Mammal -> Dog. Demonstrate multilevel inheritance.

class Animal:
    def eat(self):
        print("Animal eats food")

class Mammal(Animal):
    def walk(self):
        print("Mammal walks on land")

class Dog(Mammal):
    def bark(self):
        print("Dog barks")

d = Dog()

d.eat()
d.walk()
d.bark()
```

```python
# Q.2) Create a class Employee with class method to keep count of total employees.

class Employee:
    count = 0
    def __init__(self):
        Employee.count += 1

    @classmethod
    def show(cls):
        print("Total Employees:", cls.count)

e1 = Employee()
e2 = Employee()
e3 = Employee()
e4 = Employee()
e5 = Employee()

Employee.show()
```

```python
# Q.3) Create a class Book with special methods __str__, __len__, and __eq__.

class Book:

    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return "Book Title: " + self.title

    def __len__(self):
        return self.pages

    def __eq__(self, other):
        return self.pages == other.pages

b1 = Book("Horror", 200)
b2 = Book("Drama", 350)

print(b1)
print("Pages:", len(b1))
print("Books equal:", b1 == b2)
```

```python
# Q.4) Create a class Temperature with property decorators for Celsius and Fahrenheit CONVERSION.

class Temperature:

    def __init__(self, celsius):
        self._celsius = celsius

    @property
    def celsius(self):
        return self._celsius

    @property
    def fahrenheit(self):
        return (self._celsius * 9/5) + 32

t = Temperature(32)

print("Celsius:", t.celsius)
print("Fahrenheit:", t.fahrenheit)
```

```python
# Q.1) Create an abstract class Shape with abstract method area(). Create subclasses Rectangle,
# Circle, and Triangle that implement the area method.

from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def area(self):
        return 10 * 3

class Circle(Shape):
    def area(self):
        return 3.14 * 6 * 6

class Triangle(Shape):
    def area(self):
        return 0.5 * 6 * 3

r = Rectangle()
c = Circle()
t = Triangle()

print("Rectangle Area:", r.area())
print("Circle Area:", c.area())
print("Triangle Area:", t.area())
```

```python
# Q.2) Create a class Library that manages a collection of books.
# Implement methods to add books, remove books, search books, and display all books.

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def remove(self, book):
        self.books.remove(book)

    def search(self, book):
        print(book in self.books)

    def show(self):
        print(self.books)

l = Library()

l.add("Horror")
l.add("Drama")
l.add("Sci-fi")

l.search("Horror")
l.remove("Drama")
l.show()
```
